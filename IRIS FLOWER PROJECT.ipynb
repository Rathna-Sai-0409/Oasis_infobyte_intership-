Boxplot or Whisker plot Box plot was was first introduced in year 1969 by Mathematician John Tukey.Box plot give a statical summary of the features being plotted.Top line represent the max value,top edge of box is third Quartile, middle edge represents the median,bottom edge represents the first quartile value.The bottom most line respresent the minimum value of the feature.The height of the box is called as Interquartile range.The black dots on the plot represent the outlier values in the data.

```python
 
from sklearn.ensemble import RandomForestClassifier,VotingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
     
```

```python
 model_dict = {
    RandomForestClassifier(): {
        'n_estimators': [100, 150],
        'criterion': ['gini', 'entropy', 'log_loss'],
        'max_features': ['sqrt', 'log2', None]
    },
    SVC(): {
        'gamma': ['scale', 'auto']
    },
    LogisticRegression(): {
        'penalty': ['l1', 'l2', 'elasticnet'],
        'solver': ['lbfgs', 'liblinear', 'newton-cg', 'sag']
    },
    KNeighborsClassifier(): {
        'n_neighbors': [5, 7, 9],
        'weights': ['uniform', 'distance'],
        'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
    }
}
```

```python
for model in model_dict.keys():
    clf = GridSearchCV(estimator=model,param_grid=model_dict[model],cv=5)
    clf.fit(x_train,y_train)
    print(f"{model} best parameters{clf.best_params_},best score {clf.best_score_}")
```

```python
clf_1 = SVC(gamma="auto")
clf_2 = LogisticRegression(penalty= '12', solver= 'sag')
clf_3 = KNeighborsClassifier(algorithm="auto",n_neighbors=5, weights='uniform')
```

```python

voting_model = VotingClassifier(estimators=[('svc', clf_1), ('Lr', clf_2), ('knn', clf_3)], voting='hard')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier

# Create a LogisticRegression model with 'l2' penalty
logistic_model = LogisticRegression(penalty='l2')

# Create a VotingClassifier with the LogisticRegression model and other models
voting_model = VotingClassifier(estimators=[('lr', logistic_model), ('rf', RandomForestClassifier()), ('svc', SVC()), ('knn', KNeighborsClassifier())], voting='hard')

# Fit the ensemble model to your training data
voting_model = voting_model.fit(x_train, y_train)

# Evaluate the ensemble model on your test data
accuracy = voting_model.score(x_test, y_test)
```
